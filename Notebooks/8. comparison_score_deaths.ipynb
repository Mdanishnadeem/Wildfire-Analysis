# Comparing Wildfire Impact Score and Deaths 
## This notebook is used to compare the wildfire impact score and AQI calculated previously with the deaths caused due to Chronic Respiratory Diseases and Chronic Obstructive Pulmonary Diseases 

Importing the relevant libraries required

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
```

Importing the datasets needed for analysis 

```python
female_clean_df = pd.read_csv('/Users/apple/Desktop/DATA_512/Wildfire-Analysis/data/female_deaths.csv')
male_clean_df = pd.read_csv('/Users/apple/Desktop/DATA_512/Wildfire-Analysis/data/male_deaths.csv')
both_clean_df = pd.read_csv('/Users/apple/Desktop/DATA_512/Wildfire-Analysis/data/both_deaths.csv')
aqi_score_df = pd.read_csv('/Users/apple/Desktop/DATA_512/Wildfire-Analysis/data/aqi_score_combined.csv')
```

```python
#visualising datasets 
aqi_score_df.head()
```

```python
#visualising datasets
female_clean_df.head()
```

**Extracting and preparing the relevant data for female deaths from our dataframes for future plots**

```python
# Filter data for Chronic respiratory diseases
crd_df = female_clean_df[(female_clean_df['cause_name'] == 'Chronic respiratory diseases')]

# Select relevant columns
crd_df = crd_df[['year_id', 'mx']].reset_index(drop=True)

# Rename the 'mx' column to represent Chronic respiratory diseases deaths
crd_df = crd_df.rename(columns={'mx': 'CRD_deaths_females'})

# Filter data for Chronic respiratory diseases
copd_df = female_clean_df[(female_clean_df['cause_name'] == 'Chronic obstructive pulmonary disease')]

# Select relevant columns
copd_df = copd_df[['year_id', 'mx']].reset_index(drop=True)

# Rename the 'mx' column to represent Chronic respiratory diseases deaths
copd_df = copd_df.rename(columns={'mx': 'COPD_deaths_females'})


female_deaths_combined = pd.merge(crd_df, copd_df, on='year_id', how='outer')
```

```python
female_deaths_combined.head()
```

**Extracting and preparing the relevant data for male deaths from our dataframes for future plots**

```python
# Filter data for Chronic respiratory diseases
crd_df = male_clean_df[(male_clean_df['cause_name'] == 'Chronic respiratory diseases')]

# Select relevant columns
crd_df = crd_df[['year_id', 'mx']].reset_index(drop=True)

# Rename the 'mx' column to represent Chronic respiratory diseases deaths
crd_df = crd_df.rename(columns={'mx': 'CRD_deaths_males'})

# Filter data for Chronic respiratory diseases
copd_df = male_clean_df[(male_clean_df['cause_name'] == 'Chronic obstructive pulmonary disease')]

# Select relevant columns
copd_df = copd_df[['year_id', 'mx']].reset_index(drop=True)

# Rename the 'mx' column to represent Chronic respiratory diseases deaths
copd_df = copd_df.rename(columns={'mx': 'COPD_deaths_males'})

males_deaths_combined = pd.merge(crd_df, copd_df, on='year_id', how='outer')
```

```python
males_deaths_combined.head()
```

**Extracting and preparing the relevant data for both genders from our dataframes for future plots**

```python
# Filter data for Chronic respiratory diseases
crd_df = both_clean_df[(both_clean_df['cause_name'] == 'Chronic respiratory diseases')]

# Select relevant columns
crd_df = crd_df[['year_id', 'mx']].reset_index(drop=True)

# Rename the 'mx' column to represent Chronic respiratory diseases deaths
crd_df = crd_df.rename(columns={'mx': 'CRD_deaths_both'})

# Filter data for Chronic respiratory diseases
copd_df = both_clean_df[(both_clean_df['cause_name'] == 'Chronic obstructive pulmonary disease')]

# Select relevant columns
copd_df = copd_df[['year_id', 'mx']].reset_index(drop=True)

# Rename the 'mx' column to represent Chronic respiratory diseases deaths
copd_df = copd_df.rename(columns={'mx': 'COPD_deaths_both'})

# Display the resulting DataFrame
#result_df.head()

both_deaths_combined = pd.merge(crd_df, copd_df, on='year_id', how='outer')
```

```python
both_deaths_combined.head()
```

**Combining all the independent dataframes into one single dataframe and dropping an unwanted columns**

```python
combined_df_1 = pd.merge(aqi_score_df, female_deaths_combined, left_on='Year', right_on = 'year_id', how='inner')
```

```python
combined_df_2 = pd.merge(combined_df_1, males_deaths_combined, on='year_id', how='inner')
```

```python
final_df =  pd.merge(combined_df_2, both_deaths_combined, on='year_id', how='inner')
```

```python
final_df.drop(['Unnamed: 0', 'year_id'], axis = 1, inplace = True)
```

```python
final_df
```

**One of the row had missing values for score, to avoid any error and smooth plotting that was replaced with mean**

```python
final_df['Score'].fillna(final_df['Score'].mean(), inplace = True)
```

```python
final_df.columns
```

# HEATMAP FOR ALL OUR RELEVANT VARIABLES

```python
# Select relevant columns
selected_columns = ['AQI', 'Score', 'CRD_deaths_females', 'COPD_deaths_females',
       'CRD_deaths_males', 'COPD_deaths_males', 'CRD_deaths_both',
       'COPD_deaths_both']

# Create a subset DataFrame with selected columns
subset_df = final_df[selected_columns]

# Calculate the correlation matrix
correlation_matrix = subset_df.corr()

# Create a heatmap using seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)

plt.title('Correlation Matrix \n 1980 to 2014')
plt.show()
```

# Time Series plot to analyse trends between Score, AQI, CRD deaths & COPD deaths for both genders

```python
plt.figure(figsize=(10, 6))

# Plotting 'score' over years
plt.plot(final_df['Year'], final_df['Score'], label='Score', marker='o')

# Plotting 'crd_deaths' over years
plt.plot(final_df['Year'], final_df['CRD_deaths_both'], label='CRD Deaths', marker='o')

# Plotting 'copd_deaths' over years
plt.plot(final_df['Year'], final_df['COPD_deaths_both'], label='COPD Deaths', marker='o')

# Plotting 'AQI' over years
plt.plot(final_df['Year'], final_df['AQI'], label='AQI', marker='o')

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Line Plots for Score, CRD Deaths, and COPD Deaths Over Years \n 1980 to 2014')

# Adding legend
plt.legend()

# Display the plot
plt.show()
```

A positive relation can be observed between all of the variables
