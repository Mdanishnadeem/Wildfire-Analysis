The goal of this notebook is the make visualisations from our datasets. To try and understand the trends in fires that have occurred. 

```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_fire = pd.read_csv('/Users/apple/Desktop/DATA_512/Project/data/preprocessed_fire_data.csv', low_memory = False)
```

```python
df_fire.head()
```

# The following is a histogram showing the number of fires occurring every 50 mile distance from your assigned city up to 1250 miles.

```python
# Define the distance intervals (in miles)
distance_intervals = list(range(50, 1300, 50))

# Create a histogram of the 'distance' column
plt.figure(figsize=(10, 6))  # Set the figure size
plt.hist(df_fire[df_fire['Distance'] < 1250]['Distance'], bins=distance_intervals, edgecolor='k')

# Customize the plot
plt.title('Histogram of Fires by Distance')
plt.xlabel('Distance (miles)')
plt.ylabel('Number of Fires')
plt.grid(True)

# Show the plot
plt.show()
```

# A time series graph of total acres burned per year for the fires occurring in the specified distance from your city.

```python
# Group the data by year and calculate the sum of acres burned per year
yearly_acres_burned = df_fire.groupby('Fire_Year')['GIS_Acres'].sum()
yearly_acres_burned = yearly_acres_burned.reset_index()
yearly_acres_burned.drop(58,inplace = True)

# Create a time series plot
plt.figure(figsize=(10, 6))  # Set the figure size
plt.plot(yearly_acres_burned['Fire_Year'], yearly_acres_burned['GIS_Acres'], marker='o', linestyle='-')

# Customize the plot
plt.title('Time Series of Acres Burned per Year')
plt.xlabel('Year')
plt.ylabel('Total Acres Burned')
# Rotate the x-axis labels by 90 degrees
plt.xticks(rotation=90)
plt.grid(True)

# Show the entire y-values (disable scientific notation)
plt.ticklabel_format(style='plain', axis='y')
# Show the plot
plt.show()
```

# A time series graph containing your fire smoke estimate for your city and the AQI estimate for your city.

```python
combined_df = pd.read_csv('/Users/apple/Desktop/DATA_512/Project/data/aqi_score_combined.csv')
```

```python
# Create a line graph
plt.figure(figsize=(10, 6))

# Plot AQI data
plt.plot(combined_df['Year'], combined_df['AQI'], marker='o', label='AQI')

# Plot Score data
plt.plot(combined_df['Year'], combined_df['Score'] , marker='o', label='Score')

# Set labels and title
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('AQI and self calculated score Over the Years')

# Add a legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
