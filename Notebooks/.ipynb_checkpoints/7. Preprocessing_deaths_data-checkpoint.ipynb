# The purpose of this notebook is the preprocess the deaths dataset available and make time series visualisation

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
raw_df = pd.read_csv('/Users/apple/Desktop/DATA_512/proj_datasets/IHME_USA_COUNTY_RESP_DISEASE_MORTALITY_1980_2014_STATES_T_TO_W_CSV/Texas_data.csv')
```

```python
#To check all unique values in texas run the below code 
#raw_df['location_name'].unique()
```

```python
if len(raw_df[raw_df['location_name']=='Ector County']) > 0:
    print ('Ector County Exists')
```

```python
ector_raw_df = raw_df[raw_df['location_name']=='Ector County']
ector_raw_df.head()
```

```python
death_causes = ector_raw_df['cause_name'].unique()
print(death_causes)
```

```python
genders = ector_raw_df['sex'].unique()
print(genders)
```

```python
# Iterate over cause_names and print count of each sex type for each cause_name
for cause in death_causes:
    cause_df = ector_raw_df[ector_raw_df['cause_name'] == cause]
    sex_counts = cause_df['sex'].value_counts()
    print(f"Counts for cause_name {cause}:\n{sex_counts}\n")
```

## From the output above we can see that the count of male, female and both are same. 

```python
##Get three new dataframes for sex = both, male and female 
ector_both_raw_df = ector_raw_df[ector_raw_df['sex']=='Both']
print(f'Shape of ector_both_raw_df is {ector_both_raw_df.shape}')
ector_male_raw_df = ector_raw_df[ector_raw_df['sex']=='Male']
print(f'Shape of ector_male_raw_df is {ector_male_raw_df.shape}')
ector_female_raw_df = ector_raw_df[ector_raw_df['sex']=='Female']
print(f'Shape of ector_female_raw_df is {ector_female_raw_df.shape}')
```

# Visualisations 

## Visualisation 1 

```python
# Plot line chart for each disease type
plt.figure(figsize=(10, 6))

for cause in death_causes:
    if cause == 'Chronic respiratory diseases' or cause == 'Chronic obstructive pulmonary disease':
        cause_df = ector_both_raw_df[ector_both_raw_df['cause_name'] == cause]
        plt.plot(cause_df['year_id'], cause_df['mx'], label=cause)

plt.title('Deaths Over Time by Disease Type for both Males and Females combined \n 1980 to 2014')
plt.xlabel('Year')
plt.ylabel('Count of Deaths (mx)')
plt.legend()
plt.show()
```

## Visualisation 2 

```python
# Plot line chart for each disease type 
plt.figure(figsize=(10, 6))

for cause in death_causes:
    if cause == 'Chronic respiratory diseases' or cause == 'Chronic obstructive pulmonary disease':
        cause_df_female = ector_female_raw_df[ector_female_raw_df['cause_name'] == cause]
        cause_df_male = ector_male_raw_df[ector_male_raw_df['cause_name'] == cause]
        plt.plot(cause_df_female['year_id'], cause_df_female['mx'], label=f'females {cause}')
        plt.plot(cause_df_male['year_id'], cause_df_male['mx'], label=f'males {cause}')
        
plt.title('Deaths Over Time by Disease Type for both Male and Female separately \n 1980 to 2014')
plt.xlabel('Year')
plt.ylabel('Count of Deaths (mx)')
plt.legend()
plt.show()
```

## Visualisation 3

```python
# Filter data for the specific diseases
selected_diseases = ['Chronic respiratory diseases', 'Chronic obstructive pulmonary disease']
filtered_df = ector_female_raw_df[ector_female_raw_df['cause_name'].isin(selected_diseases)].copy(deep=True)

# Calculate the percentage change in deaths (mx) over time
filtered_df['percentage_change'] = filtered_df.groupby('cause_name')['mx'].pct_change() * 100

# Calculate the cumulative percentage change over time
filtered_df['cumulative_percentage_change'] = filtered_df.groupby('cause_name')['percentage_change'].cumsum()

# Plot the cumulative percentage change over time
plt.figure(figsize=(10, 6))

for disease in selected_diseases:
    disease_data = filtered_df[filtered_df['cause_name'] == disease]
    plt.plot(disease_data['year_id'], disease_data['cumulative_percentage_change'], label=disease)

plt.title('Cumulative Percentage Change in Deaths Over Time for Females \n 1980 to 2014')
plt.xlabel('Year')
plt.ylabel('Cumulative Percentage Change')
plt.legend()
plt.show()
```

## Visualisation 4

```python
# Filter data for the specific diseases
selected_diseases = ['Chronic respiratory diseases', 'Chronic obstructive pulmonary disease']
filtered_df = ector_male_raw_df[ector_male_raw_df['cause_name'].isin(selected_diseases)].copy(deep=True)

# Calculate the percentage change in deaths (mx) over time
filtered_df['percentage_change'] = filtered_df.groupby('cause_name')['mx'].pct_change() * 100

# Calculate the cumulative percentage change over time
filtered_df['cumulative_percentage_change'] = filtered_df.groupby('cause_name')['percentage_change'].cumsum()

# Plot the cumulative percentage change over time
plt.figure(figsize=(10, 6))

for disease in selected_diseases:
    disease_data = filtered_df[filtered_df['cause_name'] == disease]
    plt.plot(disease_data['year_id'], disease_data['cumulative_percentage_change'], label=disease)

plt.title('Cumulative Percentage Change in Deaths Over Time for Males \n 1980 to 2014')
plt.xlabel('Year')
plt.ylabel('Cumulative Percentage Change')
plt.legend()
plt.show()
```
